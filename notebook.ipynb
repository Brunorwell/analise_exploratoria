```python
import matplotlib.pyplot as plt

# filtros
df_Canada = df[df['Country/Region'] == 'Canada']
df_Us = df[df['Country/Region'] == 'United States']

# cria 1 linha e 2 colunas
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Canada
df_Canada['Ship Mode'].value_counts(normalize=True).plot(
    kind='bar',
    ax=axes[0],
    title='Canada'
)

# USA
df_Us['Ship Mode'].value_counts(normalize=True).plot(
    kind='bar',
    ax=axes[1],
    title='United States'
)

plt.tight_layout()
plt.show()
```

```python
# A média não é um bom indicador. Notamos que o desvio padrão é alto demais comparado com a mediana.

def resumo_sales(df, group_col, value_col='Sales'):
    resumo = df.groupby(group_col)[value_col].describe()
    
    resumo['IQR'] = resumo['75%'] - resumo['25%']
    resumo['RCVq'] = (0.75 * resumo['IQR']) / resumo['50%']
    resumo['QCD'] = (resumo['75%'] - resumo['25%']) / (resumo['75%'] + resumo['25%'])
    resumo['CV (%)'] = (resumo['std'] / resumo['mean']) * 100
    
    return resumo

resumo_canada = resumo_sales(df_Canada, 'Ship Mode')
resumo_us = resumo_sales(df_Us, 'Ship Mode')

resumo_canada['Country'] = 'Canada'
resumo_us['Country'] = 'United States'

df_resumo_final = pd.concat([resumo_canada, resumo_us])

df_resumo_final[['Country', 'min','max','mean', 'std', 'CV (%)', '25%', '50%', '75%', 'IQR', 'RCVq', 'QCD']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = df['Ship Mode'].value_counts().index

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# função pra calcular métricas
def calc_metrics(df, group_col, value_col):
    g = df.groupby(group_col)[value_col]
    q1 = g.quantile(0.25)
    q3 = g.quantile(0.75)
    median = g.median()
    
    iqr = q3 - q1
    rcvq = (0.75 * iqr) / median
    qcd = (q3 - q1) / (q3 + q1)
    limit_superior = q3 + (1.5 * iqr)
    
    return rcvq, qcd, limit_superior

# -------- Canada --------
sns.boxplot(data=df_Canada, x='Ship Mode', y='Sales', ax=axes[0], order=order)
axes[0].set_title('Canada')

rcvq_can, qcd_can, limit_superior_can = calc_metrics(df_Canada, 'Ship Mode', 'Sales')
rcvq_can = rcvq_can.reindex(order)
qcd_can = qcd_can.reindex(order)

axes[0].set_ylim(0, limit_superior_can.max())

for i, mode in enumerate(rcvq_can.index):
    axes[0].text(
        i, 450,  # posição X = índice da categoria
        f'QCD: {qcd_can[mode]:.2f}\nRCVq: {rcvq_can[mode]:.2f}',
        ha='center',
        fontsize=8,
        bbox=dict(facecolor='white', alpha=0.6)
    )

# -------- USA --------
sns.boxplot(data=df_Us, x='Ship Mode', y='Sales', ax=axes[1], order=order)
axes[1].set_title('United States')

rcvq_us, qcd_us, limit_superior_us = calc_metrics(df_Us, 'Ship Mode', 'Sales')
rcvq_us = rcvq_us.reindex(order)
qcd_us = qcd_us.reindex(order)

axes[1].set_ylim(0, limit_superior_us.max())


for i, mode in enumerate(rcvq_us.index):
    axes[1].text(
        i, 450,
        f'QCD: {qcd_us[mode]:.2f}\nRCVq: {rcvq_us[mode]:.2f}',
        ha='center',
        fontsize=8,
        bbox=dict(facecolor='white', alpha=0.6)
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = df['Ship Mode'].value_counts().index

# países
dfs = {
    'Canada': df_Canada,
    'United States': df_Us
}

fig, axes = plt.subplots(
    nrows=2,
    ncols=len(order),
    figsize=(16, 8),
    sharex=True,
    sharey=True
)

for row_idx, (country, df_country) in enumerate(dfs.items()):
    for col_idx, mode in enumerate(order):
        
        ax = axes[row_idx, col_idx]
        
        subset = df_country[df_country['Ship Mode'] == mode]
        
        sns.histplot(
            data=subset,
            x='Sales',
            bins=20,
            ax=ax
        )
        
        # título da coluna (Ship Mode)
        if row_idx == 0:
            ax.set_title(mode)
        
        # label da linha (país)
        if col_idx == 0:
            ax.set_ylabel(country)
        else:
            ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

### Conclusão

Em todas as variáveis qualitativas nominais analisadas, observa-se uma distribuição assimétrica à direita. Nesses casos, medidas como média e desvio padrão tornam-se menos representativas, pois possuem **ponto de ruptura igual a 0% - Essas medidas não suportam sequer um valor extremo**

Diante disso, utilizamos medidas mais robustas, como o Coeficiente Robusto de Variação (RCV) e o Coeficiente de Dispersão Quartílica (QCD), que são baseados nos quartis e apresentam maior resistência a outliers, **com ponto de ruptura igual a 25%, o que significa que elas suportam até 25% de ruidos até serem impactadas por outliers**.

A análise do RCV indica que, mesmo considerando apenas os 50% dos dados, há alta variabilidade em todos os agrupamentos. Em média, a dispersão dos valores centrais é cerca de 2.38 vezes o valor da mediana, evidenciando grande variabilidade nos dados.

Dessa forma, não é possível identificar uma diferença significativa na consistência dos valores de vendas entre as variáveis categóricas analisadas (Ship Mode, Segment e Category). Em geral, os dados apresentam alta variabilidade e baixa consistência, independentemente do agrupamento.
